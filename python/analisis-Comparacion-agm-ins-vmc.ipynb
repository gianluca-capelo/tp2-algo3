```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def correlacion_pearson(data_frame, columna1, columna2):
    return np.corrcoef(data_frame[columna1], data_frame[columna2])[0,1];
```

# Métodos
Los métodos evaluados en este notebook son los siguientes:

- **VMC** : vecino mas cercano
- **AGM** : arbol generador minimo
- **I** : incersion

# Experimentos
Leemos el dataframe con todos los resultados.

```python
df_resultados = pd.read_csv("resultados/resultado.csv");
df_resultados
```

```python
df_resultados["metodo"].unique()
```

## agm vs ins vs vmc

```python
plt.ylim(0, 15000)
df = df_resultados.sort_values(by=['costo'],ascending=True)
df = df[(df["dataset"] == "dataset_aleatorias_agm_ins")].copy();
sns.set_style("darkgrid")
ax = sns.barplot(x="n", y="costo", hue="metodo", data=df, palette="flare")
plt.legend(loc ='upper left')
plt.savefig("graficos/vmc-ins-agm.svg")
```

## agm vs ins vs vmc EUCLIDIANO

```python
plt.ylim(0, 15000)
df = df_resultados.sort_values(by=['metodo'],ascending=False)
df = df[(df["dataset"] != "dataset_aleatorias_agm_ins")].copy();
sns.set_style("darkgrid")
ax = sns.barplot(x="n", y="costo", hue="metodo", data=df, palette="flare")
plt.legend(loc ='upper left')
plt.savefig("graficos/vmc-ins-agm-euc.svg")
```
